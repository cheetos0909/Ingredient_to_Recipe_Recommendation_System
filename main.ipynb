 The Function **has_hindi()** checks whether a given text string contains characters from the Hindi script (Devanagari). It returns True, if at least one character in the text falls within the specified Unicode range for Hindi script, and it returns False, if the input is not a string or if no characters from the Hindi script are present.

```python
def has_hindi(text):
    """ This Function checks whether a given text string contains characters from the Hindi script (Devanagari). It returns True, if at least one character in the text falls within the specified Unicode range for Hindi script, and it returns False, if the input is not a string or if no characters from the Hindi script are present."""
    if isinstance(text, str):
        hindi_range_bool = any(0x0900 <= ord(c) <= 0x097F for c in text)
        return hindi_range_bool
    return False
```

The **remove_punctuations_except()** function in Python takes a list of strings as input and removes all punctuation marks from each string, except for the ones specified in the exceptions parameter. It returns a list of cleaned strings with the specified exceptions.

```python
def remove_punctuations_except(input_list,exceptions=['(',')',',','/']):
  """The function in Python takes a list of strings as input and removes all punctuation marks from each string, except for the ones specified in the exceptions parameter. It returns a list of cleaned strings with the specified exceptions."""
  cleaned_list=[]
  for element in input_list:
    cleaned_element=''.join(char for char in element if char  in exceptions or char not in string.punctuation)
    cleaned_list.append(cleaned_element)
  return cleaned_list
```

### **Combining above pre-processing functionalities into below declared function in order to apply all  functionalities on all cells in one go**

```python
def process_to_dictionary(text):


  def remove_punctuations_except(input_list,exceptions=['(',')',',','/']):
    cleaned_list=[]
    for element in input_list:
      cleaned_element=''.join(char for char in element if char  in exceptions or char not in string.punctuation)
      cleaned_list.append(cleaned_element)
    return cleaned_list

  text_list=remove_punctuations_except(text)


  #Create a dictionary
  Ingredients = {}

  for item in text_list:
    words = item.split()
    # print(words)
    quantity=""
    ingredients=""
    for word in words:
      if word in quantity_list:
        quantity += word + " "
      else:
        ingredients += word + " "
    quantity = quantity.strip()  # Remove leading/trailing spaces
    ingredients = ingredients.strip()  # Remove leading/trailing spaces

    if quantity and ingredients:
      Ingredients[ingredients] = quantity

  return Ingredients



```

# **Taking Input From User And Returning Sorted Recommendations**

### **Working of Recommendation System :**
#### 1)Local Ingredient : Ingredient Present with Consumer at Home
#### 2)Total Ingredient : Total Ingredients required to make Recipe
#### 3)Foreign Ingredient (Also Known as Ingredient Difference): Total Ingredient-Local Ingredient
#### Lesser the Ingredient Difference , higher the chance recipe would be recommended

```python
import re
target_ingredients_input = input("Enter target ingredient names separated by commas: ")
target_ingredients = [ingredient.strip() for ingredient in target_ingredients_input.split(',')]
def contains_all_target_ingredients(ingredient_dict):
    ingredient_text = ' '.join(ingredient_dict.keys())
    return all(re.search(ingredient, ingredient_text, re.IGNORECASE) for ingredient in target_ingredients)
filtered_rows = dataset2[dataset2['processed_ingredients'].apply(contains_all_target_ingredients)]
filtered_rows['ingredient_difference'] = filtered_rows['processed_ingredients'].apply(lambda x: len(x) - len(target_ingredients))
```
